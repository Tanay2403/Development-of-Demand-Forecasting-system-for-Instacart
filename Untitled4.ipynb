```python
import pandas as pd
```

```python
orders = pd.read_csv(r'D:\abhijeet\project\archive\orders.csv')
departments = pd.read_csv(r'D:\abhijeet\project\archive\departments.csv')
prior_order = pd.read_csv(r'D:\abhijeet\project\archive\order_products__prior.csv')
aisles = pd.read_csv(r'D:\abhijeet\project\archive\aisles.csv')
train_order = pd.read_csv(r'D:\abhijeet\project\archive\order_products__train.csv')
products = pd.read_csv(r'D:\abhijeet\project\archive\products.csv')
```

```python
#concat train order and prior orders
prior_train_orders = pd.concat([prior_order, train_order]).sort_values(by=['order_id'])

#merge with rest
prior_train_orders = pd.merge(left = prior_train_orders, right = products,
                             left_on='product_id', right_on='product_id').sort_values(by=['order_id']).reset_index(drop=True)
prior_train_orders = pd.merge(left = prior_train_orders, right = aisles,
                             left_on='aisle_id', right_on='aisle_id').sort_values(by=['order_id']).reset_index(drop=True)
prior_train_orders = pd.merge(left = prior_train_orders, right = departments,
                             left_on='department_id', right_on='department_id').sort_values(by=['order_id']).reset_index(drop=True)
prior_train_orders = pd.merge(left = prior_train_orders, right = orders,
                             left_on='order_id', right_on='order_id').sort_values(by=['order_id']).reset_index(drop=True)

col_order = ['user_id','order_id','product_id','aisle_id','department_id','add_to_cart_order',
 'reordered','product_name','aisle','department','eval_set','order_number','order_dow','order_hour_of_day',
 'days_since_prior_order']

prior_train_orders = prior_train_orders[col_order] 
```

```python
print(prior_train_orders.isnull().sum())
```

```python
fraction_to_keep = 0.98  # Adjust the fraction as needed
prior_train_orders2 = prior_train_orders.sample(frac=fraction_to_keep)
```

```python
print(len(prior_train_orders2))
```

```python
# Drop rows with missing values
prior_train_orders = prior_train_orders2.dropna()

# Impute missing values
# df = df.fillna(some_value)
print(prior_train_orders2.isnull().sum())
```

```python
reorder_rates = prior_train_orders2.groupby('product_id')['reordered'].mean()
prior_train_orders2['reorder_rate'] = prior_train_orders2['product_id'].map(reorder_rates)
```

```python
import statsmodels.api as sm

# Step 3: Model Selection
# Example: (p,d,q) = (2, 1, 1) and (P,D,Q,s) = (1, 1, 1, 7)
# You need to adjust these parameters based on your data and analysis
order_params = (2, 1, 1)
seasonal_order_params = (1, 1, 1, 7)

# Step 4: Model Fitting
# Define SARIMA model
sarima_model = sm.tsa.SARIMAX(endog=prior_train_orders2['add_to_cart_order'],
                               order=order_params,
                               seasonal_order=seasonal_order_params,
                               enforce_stationarity=False,
                               enforce_invertibility=False)

# Fit SARIMA model
sarima_results = sarima_model.fit()

# Step 5: Model Evaluation
# Example: Evaluate model performance using AIC
print("AIC:", sarima_results.aic)

# Step 6: Forecasting
# Forecast future values
# Example: Forecast next 10 time points
forecast = sarima_results.forecast(steps=10)

print("Forecasted values:", forecast)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Actual values (ground truth)
actual_values = prior_train_orders2['add_to_cart_order']

# Predicted values from SARIMA model
predicted_values = sarima_results.predict()

# Calculate metrics
mse = mean_squared_error(actual_values, predicted_values)
mae = mean_absolute_error(actual_values, predicted_values)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((actual_values - predicted_values) / actual_values)) * 100

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Percentage Error (MAPE):", mape)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Load and preprocess your data
# (Assuming you have already loaded and preprocessed your data as 'prior_train_orders1')

# Perform feature engineering if needed
# For example, you can create lagged variables or include additional relevant features

# Split the data into train and test sets
train_size = int(0.97 * len(prior_train_orders2))
train_data = prior_train_orders2[:train_size]
test_data = prior_train_orders2[train_size:]

# Model Selection and Parameter Tuning
best_aic = np.inf
best_order = None
best_seasonal_order = None

for p in range(3):  # Range for p
    for d in range(2):  # Range for d
        for q in range(3):  # Range for q
            for P in range(2):  # Range for seasonal P
                for D in range(2):  # Range for seasonal D
                    for Q in range(2):  # Range for seasonal Q
                        try:
                            model = SARIMAX(train_data['add_to_cart_order'],
                                            order=(p, d, q),
                                            seasonal_order=(P, D, Q, 7),
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
                            results = model.fit()
                            if results.aic < best_aic:
                                best_aic = results.aic
                                best_order = (p, d, q)
                                best_seasonal_order = (P, D, Q, 7)
                        except:
                            continue

print("Best SARIMA Order:", best_order)
print("Best SARIMA Seasonal Order:", best_seasonal_order)

# Fit the SARIMA model with the best parameters
final_model = SARIMAX(train_data['add_to_cart_order'],
                      order=best_order,
                      seasonal_order=best_seasonal_order,
                      enforce_stationarity=False,
                      enforce_invertibility=False)
final_results = final_model.fit()

# Forecast future values
start_index = len(train_data)
end_index = start_index + len(test_data) - 1
predicted_values = final_results.predict(start=start_index, end=end_index)

# Evaluate the model on the test set
mse = mean_squared_error(test_data['add_to_cart_order'], predicted_values)
mae = mean_absolute_error(test_data['add_to_cart_order'], predicted_values)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)

# Visualize the actual vs. predicted values
plt.figure(figsize=(10, 6))
plt.plot(test_data.index, test_data['add_to_cart_order'], label='Actual')
plt.plot(test_data.index, predicted_values, label='Predicted', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Number of Items Added to Cart')
plt.title('Actual vs. Predicted Values')
plt.legend()
plt.grid(True)
plt.show()
```
